```python
import pickle
import matplotlib.pyplot as plt


with open('20230217_daily_data.pickle', 'rb') as handle:
    my_dict = pickle.load(handle)

len(my_dict['221840'])
```

```python
dict_keys_list = list(my_dict.keys())
```

```python
def price_plot(num):

    data = my_dict[dict_keys_list[num]]
    plt.plot(my_dict[dict_keys_list[num]])
    plt.show()

    return data
```

```python
# draw a current position and the threshold as a line 


data = price_plot(3)

```

```python
idx = 0
sell_percent = 0.99
threshold = data[idx]*sell_percent
high_so_far = data[idx]
sold_idx = []
stock_in_hand = True


for idx in range(len(data)):


    if high_so_far < data[idx]:
        high_so_far = data[idx]
        threshold = data[idx]*sell_percent
    else:
        if data[idx] < threshold and stock_in_hand:
            sold_idx.append(idx)
            print('It was sold in ' + str(idx) + '.')
            stock_in_hand = False


fig, ax = plt.subplots()
ax.plot(data)
ax.scatter(idx, data[idx], marker='*', c='r', s=100)

if len(sold_idx) > 0:
    ax.scatter(sold_idx[0], data[sold_idx[0]], marker='*', c='b', s=100)
else:
    pass

ax.axhline(y=threshold, color='red', linestyle='--')
plt.show()

len(sold_idx)
```

```python

def plot_current_time(idx):
    
    idx = int(idx)
    
    sell_percent = 0.99

    fig, ax = plt.subplots()
    threshold = data[idx]*sell_percent
    ax.plot(data)
    ax.scatter(idx, data[idx], marker='*', c='r', s=100)
    ax.axhline(y=threshold, color='red', linestyle='--')
    plt.show()
```
