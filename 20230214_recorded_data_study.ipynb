```python
import numpy as np

def convolve_last_n(values, n):
    if len(values) < n:
        values[-1]
        #raise ValueError("The number of values must be greater than or equal to n.")
    kernel = np.ones(n) / n
    return np.convolve(values, kernel, mode='valid')
```
