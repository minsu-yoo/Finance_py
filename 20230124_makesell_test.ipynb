```python
def price_data_gen(*args):

    len_args = len(args)
    com_prices = {}

# Initialize starting price

    for code in args:

        starting_price = 100

        # Generate a list of pseudo prices for a day
        prices = [starting_price]
        for i in range(1, 21600):
            price_change = random.uniform(-0.03, 0.03)
            prices.append(prices[i-1] * (1 + price_change))

        com_prices[code] = prices

    return com_prices
```
