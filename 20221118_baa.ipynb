```python
import yfinance as yf
import matplotlib.pyplot as plt
import pickle 

```

```python
spy = yf.download('SPY', start='2020-11-01', end = '2022-10-01',
           progress = False,
           auto_adjust = False, 
           actions = 'inline')
```

```python
# 최근 12개월간의 momentum score 구하기 

spy['close_1m'] = (spy['Close']/spy['Close'].shift(20)-1)*12
spy['close_3m'] = (spy['Close']/spy['Close'].shift(60)-1)*4
spy['close_6m'] = (spy['Close']/spy['Close'].shift(120)-1)*2
spy['close_12m'] = (spy['Close']/spy['Close'].shift(240)-1)*1

spy['score'] = spy['close_1m'] + spy['close_3m'] + spy['close_6m'] + spy['close_12m']
```

```python
spy['momentum_score'] = ((spy['Close']/spy['Close'].shift(20)-1)*12 + (spy['Close']/spy['Close'].shift(60)-1)*4 +
(spy['Close']/spy['Close'].shift(120)-1)*2 + (spy['Close']/spy['Close'].shift(240)-1)*1)
```

```python
# momentum score 를 만들 수 있는 function

def m_score(asset_data):
    
    result = ((spy['Close']/spy['Close'].shift(20)-1)*12 + (spy['Close']/spy['Close'].shift(60)-1)*4 +
(spy['Close']/spy['Close'].shift(120)-1)*2 + (spy['Close']/spy['Close'].shift(240)-1)*1)
    
    return result
    
    
```
