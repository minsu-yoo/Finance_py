```python
import yfinance as yf
import matplotlib.pyplot as plt
import pickle 
```

```python
spy = yf.download('SPY', start='1970-01-01', end = '2022-09-07',
           progress = False,
           auto_adjust = False, 
           actions = 'inline')
```

```python
ief = yf.download('IEF', start='1970-01-01', end = '2022-09-07',
           progress = False,
           auto_adjust = False, 
           actions = 'inline')
```

```python
tlt = yf.download('TLT', start='1970-01-01', end = '2022-09-07',
           progress = False,
           auto_adjust = False, 
           actions = 'inline')
```

```python
dbc = yf.download('DBC', start='1970-01-01', end = '2022-09-07',
           progress = False,
           auto_adjust = False, 
           actions = 'inline')
```

```python
gld = yf.download('GLD', start='1970-01-01', end = '2022-09-07',
           progress = False,
           auto_adjust = False, 
           actions = 'inline')
```

```python
a4s_data = {}
```

```python
#all four seasons data 

a4s_data['spy'] = spy
a4s_data['bond_ief'] = ief
a4s_data['long_bond_tlt'] = tlt
a4s_data['gold_gld'] = gld
a4s_data['material_dbc'] = dbc



```

```python
with open('allSeason.pickle', 'wb') as handle:
    pickle.dump(a4s_data, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
plt.style.use('plotstyle.mplstyle')

```

```python
def grow_rate(data, i, date):



    keys_list = list(data)  
    #you can get the names of keys

    #print(keys_list[i])


    growth_rate  = (data[keys_list[i]].Close.loc[date:] 
          / data[keys_list[i]].Close.loc[date])*100
    plt.plot(growth_rate, label= keys_list[i])


    return growth_rate

```

```python
gr = grow_rate(a4s_data, 0, '2010-01-11')
```

```python
growth_rate_list = []
asset_keys_list = list(a4s_data) 


growth_rate = {}

for i in range(5):
    
    a = grow_rate(a4s_data, i,'2009-01-08')
    growth_rate_list.append(a)
    growth_rate[asset_keys_list[i]] = a
    
plt.legend()
plt.show()
```
